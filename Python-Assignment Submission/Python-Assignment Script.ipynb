### 1. Write a Python program to calculate number of days between two dates. Hint: use Datetime package/module.
Sample dates : (2014, 7, 2), (2014, 7, 11)
Expected output : 9 days


```python
from datetime import datetime
date_format = "%Y/%m/%d"
first_Date = datetime.strptime('2014/7/2', date_format)
last_Date = datetime.strptime('2014/7/11',date_format)
delta = last_Date - first_Date
print(delta.days,"days")
```

### 3. Write a Python program to count the number 4 in a given list.
List = [1,4,6,8,4,9,4]


```python
def count_4_list(nums):
  count = 0
  for num in nums:
    if num == 4:
      count = count + 1
    
  return count

print(count_4_list([1, 4, 6, 8, 4, 9, 4]))
```
